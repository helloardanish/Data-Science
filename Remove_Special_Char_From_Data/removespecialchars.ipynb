```python
dfcolumns = df.columns.tolist()
```

```python
dfcolumns
```

```python
for index, row in df.iterrows():
  for col in dfcolumns:
    print(row[col])
```

```python
spccharlst = ['*', '$', '@', '-', '%', '#', '&']
for index, row in df.iterrows():
  for col in dfcolumns:
    temp = ""
    celldata = row[col]
    for chr in celldata:
      if chr in spccharlst:
        continue
      else:
        temp += chr
    print(temp)
    row[col] = temp
```
